```python
from sklearn import tree
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
import graphviz 
from numpy import random
# seleccionamos un numero random entre 1 y 1000 
r = random.randint(1, 1000)

# usamos ese numero aleatorio como seed
np.random.seed(r)


# carga de digitos
digits = load_digits()

# divicion datos 70-30%
Xt, Xv, yt, yv = train_test_split(digits.data, digits.target, test_size=0.3)


# creacion de arbol 1 y 2 
arbol_decision1 = tree.DecisionTreeClassifier(criterion = "entropy", max_depth=2)
arbol_decision2 = tree.DecisionTreeClassifier(criterion = "entropy", max_depth=5)

# creacion de arbol 3 y 4 
arbol_decision3 = tree.DecisionTreeClassifier(criterion = "entropy", min_impurity_decrease = 0.1 )
arbol_decision4 = tree.DecisionTreeClassifier(criterion = "entropy", min_impurity_decrease = 0.4 )

# creacion de arbol completo por entropia 
arbol_decision5 = tree.DecisionTreeClassifier(criterion = "entropy")

# generacion arboles 1 y 2 con datos de entrenamiento o prueba  
arbol1 = arbol_decision1.fit(Xt, yt)
arbol2 = arbol_decision2.fit(Xt, yt)

# generacion arboles 3 y 4 con datos de entrenamiento o prueba 
arbol3 = arbol_decision3.fit(Xt, yt)
arbol4 = arbol_decision4.fit(Xt, yt)

# generacion de arbol completo con datos de entrenamiento o prueba 
arbol5 = arbol_decision5.fit(Xt, yt)

#obtenemos el desempeño del sistema 
acurracy = arbol_decision1.score(Xv, yv)
print("el desempeño del sistema es de:",round(acurracy,4),"--->", round(acurracy*100,2),"%")
```

```python
#graficamos el arbol 1 
dot_data = tree.export_graphviz(arbol1, out_file=None,filled=True,rounded=True,node_ids=True)
graph = graphviz.Source(dot_data)
graph
```

```python
#graficamos el arbol 2
dot_data = tree.export_graphviz(arbol2, out_file=None,filled=True,rounded=True,node_ids=True) 
graph = graphviz.Source(dot_data) 
graph
```

```python
#graficamos el arbol 3
dot_data = tree.export_graphviz(arbol3, out_file=None,filled=True,rounded=True,node_ids=True) 
graph = graphviz.Source(dot_data) 
graph
```
